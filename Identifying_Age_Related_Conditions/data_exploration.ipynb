```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
train = pd.read_csv('./data/train.csv')
test = pd.read_csv('./data/test.csv')
greeks = pd.read_csv('./data/greeks.csv')
sample_submission = pd.read_csv('./data/sample_submission.csv')
```

```python
for col in train.columns:
    if col in ['Id', 'Class']:
        continue

    fig = go.Figure()
    fig.add_trace(go.Histogram(x=train[col], name=f'Feature {col}'))
    fig.update_layout(xaxis_title_text='Value', yaxis_title_text='Count')
    fig.write_image(f'features_distrs/{col}.png', scale=2)
```

```python
train_0 = train.loc[train.Class == 0]
train_1 = train.loc[train.Class == 1]
train_0.shape, train_1.shape
```

```python
for col in train.columns:
    if col in ['Id', 'Class']:
        continue

    fig = go.Figure()
    fig.add_trace(go.Histogram(x=train_0[col], name=f'Class 0', nbinsx=20, histnorm='probability'))
    fig.add_trace(go.Histogram(x=train_1[col], name=f'Class 1', nbinsx=20, histnorm='probability'))
    fig.update_layout(barmode='overlay', xaxis_title_text='Value', yaxis_title_text='Probability')
    fig.update_traces(opacity=0.75)
    fig.write_image(f'features_distrs_by_classes/{col}.png', scale=2)
```
